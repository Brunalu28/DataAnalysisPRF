***Unindo todos os meses de 2021 em um só dataframe:***

Foi usado o concat pra fazer a união, o axis e o ignore_index para evitar um valor duplicado no índice.

```python
anualdf = pd.concat([janDF, fevDF, marDF, abrDF, maiDF, junDF, julDF, agoDF, setDF, outDF, novDF, dezDF], axis=0, ignore_index=True)

anualdf
```


###**2ª Tarefa**

Explore os conjuntos de dados, respondendo as questões:


***(a) quantas infrações ocorreram em cada ano?***

No caso, foi verificada a quantidade de infrações por mês, com base na quantidade de infrações, primeiro foi usado o datetime para converter a string em objeto datatime e o argumento dayfirst para ser analisado primeiro os dias. Pra pegar somente os meses, foi usado o dt.month e depois contalbilizados com o count.

```python
anualdf["Data da Infração (DD/MM/AAAA)"] = pd.to_datetime(anualdf["Data da Infração (DD/MM/AAAA)"], dayfirst=True)
anualdf["infraMes"] = anualdf["Data da Infração (DD/MM/AAAA)"].dt.month
totalInfra = anualdf.groupby("infraMes")["Qtd Infrações"].count()
totalInfra
```

***(c) quais as cinco cidades brasileiras onde mais ocorreram infrações em rodovias federais?***

Foi usado o groupby para agrupar as infrações por município e logo em seguida o count, como está sendo pedido os cinco municípios com mais infrações, a lista foi ordenada de forma inversa e depois exposto os cinco primeiros municípios da lista.

```python
cidades = anualdf.groupby("Município")["Qtd Infrações"].count()
listacidades = cidades.sort_values(ascending=False)
listacidades.head(5)
```

***(a) ranking de infrações por estado;***

Para fazer o ranking das infrações por estado primeiro foram contabilizadas as infrações, depois ordenadas com o rank que classificou as ufs, em seguida foi feito o dataframe, renomeado o título e nome da coluna.

```python
rankingUF = anualdf["UF Infração"].value_counts()
rankingUF.rank
dfRanking = pd.DataFrame(rankingUF)
dfRanking.rename(columns={'UF':'Infrações'}, inplace=True)
dfRanking.index.names = ['Municipios']
dfRanking
```

***(b) infrações por dia da semana para o ano inteiro***

Primeiro foi pego o dataframe anual e a coluna data da infração foi convertida em um objeto tipo data, para facilitar a análise apenas dos dias da semana.

Depois foi usado o módulo locale para converter a localidade dos dados de data.

em seguida foi feito o dataframe, renomeado o título e nome da coluna.

```python
import locale
from datetime import datetime

locale.setlocale(locale.LC_ALL, locale= 'pt_BR.utf8')
#!sudo dpkg-reconfigure locales

infraData = anualdf
infraData["Data da Infração"] = pd.to_datetime(infraData["Data da Infração (DD/MM/AAAA)"])
infraData.set_index("Data da Infração", inplace=True)

dias = infraData.groupby(anualdf.index.day_name(locale="pt_BR.utf8"))["Qtd Infrações"].sum().sort_values(ascending=False)

dfDiasDaSemana = pd.DataFrame(dias)
dfDiasDaSemana
```

Por fim foi criado o csv. 

Conferindo o arquivo csv dos dias da semana:

```python
dfDiasDaSemana.to_csv("DiasDaSemana.csv", index=True)
DiasDaSemana = pd.read_csv("DiasDaSemana.csv")
DiasDaSemana
```
